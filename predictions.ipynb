```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
```

```python
with open('gender_encoder.pkl', 'rb') as pkl:
    gender_encoder = pickle.load(pkl)
with open('geography_encoder.pkl', 'rb') as pkl:
    geography_encoder = pickle.load(pkl)
with open('scaler.pkl', 'rb') as pkl:
    scaler = pickle.load(pkl)
model = load_model('model.h5')
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded = geography_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns = geography_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df = pd.concat([input_df, geo_encoded_df], axis = 1).drop(['Geography'], axis = 1)
input_df
```

```python
input_df['Gender'] = gender_encoder.transform(input_df['Gender'])
input_df
```

```python
scaled_df = scaler.transform(input_df)
scaled_df
```

```python
prediction = model.predict(scaled_df)
if prediction[0][0] > 0.5:
    print('Likely to churn')
else:
    print("Will not Churn")
```
